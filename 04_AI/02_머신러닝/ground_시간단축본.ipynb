```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
ground_value = pd.read_csv('2022 표준지 공시지가.csv', sep=',', header=0)
ground_value.head()
```

```python
ground2 = ground_value.copy()
```

```python
ground_data = pd.read_csv('ground.csv', sep=',', header=0, encoding = 'cp949')
ground_data.head()
```

```python
ground3 = pd.concat([ground_data, ground2['공시지가']], axis = 1)
```

```python
ground3
```

```python
target = ground3['공시지가']
```

```python
ground_data = ground3[ground3.columns.difference(['공시지가','지리적위치1_2','지리적위치2_2'])]
```

```python
ground_standardized = (ground_data - ground_data.mean()) / ground_data.std()
```

```python
ground_data_standardized = pd.concat([target, ground_standardized], axis = 1)
```

```python
ground_data_standardized
```

```python
from statsmodels.formula.api import ols
my_formula = '공시지가 ~ 시도명 + 시군구명 + 지목 + 면적 + 용도지역1 + 용도지역2 + 이용상황 + 주위환경 + 지세명 + 형상명 + 도로교통 + 방위'
lm = ols(my_formula, data=ground_data_standardized).fit()
lm.summary()
```

```python
ground_data = pd.concat([ground3['시도명'],ground3['시군구명'], ground3['지목'], ground3['면적'],
                        ground3['용도지역2'], ground3['이용상황'],ground3['주위환경'],
                        ground3['지세명'],ground3['형상명'], ground3['도로교통']
                        ,ground3['방위']], axis =1)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(ground_data, target, random_state=42)
```

```python
poly = PolynomialFeatures(include_bias=False)

poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
```

# 규제전 표준화

```python
ss = StandardScaler() # 표준화 작업을 함수로서 지원
ss.fit(train_poly)

train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

```python
lasso = Lasso(0.001)
lasso.fit(train_scaled, train_target)
print(f'training_score : {lasso.score(train_scaled, train_target)}')
print(f'test_score : {lasso.score(test_scaled, test_target)}')
```
