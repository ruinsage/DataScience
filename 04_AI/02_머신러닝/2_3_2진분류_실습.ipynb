# Step1] 심장질환 예측
* 데이터셋: heart.csv
* Mission
 - 앞에 배운 내용을 참고로 하여 Feature, label(Target)을 분류
 - 훈련데이터셋, 테스트 데이터셋 분류
 - 데이터 표준화
 - 모델 성능 평가

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
heart = pd.read_csv('heart.csv', sep=',', header=0)
heart.head()
```

```python
heart.shape
```

```python
heart_data = pd.concat([heart['sex'], heart['cp'], heart['trestbps'], heart['restecg'],
              heart['thalach'], heart['oldpeak'], heart['ca'], heart['thal']], axis = 1)
```

```python
heart_data
```

```python
heart_target = heart['target']
```

```python
heart_target
```

```python
len(heart)
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    heart_data, heart_target, stratify = heart_target, random_state=100)
```

```python
train_input.shape
```

```python
train_target.shape
```

```python
test_input.shape
```

```python
test_target.shape
```

```python
mean = np.mean(train_input, axis = 0)
std = np.std(train_input, axis = 0)
```

```python
print(mean, std)
```

```python
train_scaled = (train_input - mean) / std
```

```python
test_scaled = (test_input - mean) / std
```

```python
train_scaled
```

```python
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_target)
kn.score(train_input, train_target)
```

```python
kn.score(test_scaled, test_target)
```

# Step2] 펭귄 종류 예측
* Seaborn에 있는 펭귄데이터에 펭귄 종을 예측
 - 2개 종을 예측 할 수 있도록 데이터 전처리 및 모든 조합의 모델 생성  
   (Adelie, Chinstrap, Gentoo)
   - Adelie, Chinstrap
   - Adelie, Gentoo
   - Chinstrap, Gentoo

```python
import seaborn as sns
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
penguins2 = penguins.copy()
```

```python
penguins2
```

```python
penguins2.replace('Adelie',1, inplace =True)
penguins2.replace('Gentoo',2, inplace =True)
penguins2.replace('Chinstrap',0, inplace =True)
penguins2.replace('Torgersen',0, inplace =True)
penguins2.replace('Biscoe',1, inplace =True)
penguins2.replace('Dream',2, inplace =True)
penguins2.replace('Male',1, inplace =True)
penguins2.replace('Female',0, inplace =True)
```

```python
penguin2 = penguins2.dropna(inplace=True)
```

```python
penguin2
```

```python
Adelie_penguins = penguins2.loc[penguins['species']== 'Adelie',:]
Gentoo_penguins = penguins2.loc[penguins['species']== 'Gentoo',:]
Chinstrap_penguins = penguins2.loc[penguins['species']== 'Chinstrap',:]
```

```python
Adelie_penguins
```

```python
Gentoo_penguins
```

# Adelie_Gentoop

```python
Adelie_Gentoop = pd.concat([Adelie_penguins,Gentoo_penguins])
Adelie_Gentoop
```

```python
penguins_data = Adelie_Gentoop[Adelie_Gentoop.columns.difference(['species'])]
penguins_target = Adelie_Gentoop['species']
```

```python
penguins_data
```

```python
penguins_target
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    penguins_data, penguins_target, stratify=penguins_target, random_state=200)
```

```python
train_input
```

```python
test_input
```

```python
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
```

```python
mean
```

```python
std
```

```python
train_scaled=(train_input-mean)/std
test_scaled = ((test_input) - mean) / std
```

```python
train_scaled
```

```python
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_target)
kn.score(train_scaled, train_target)
```

```python
kn.fit(train_scaled, train_target)
kn.score(test_scaled, test_target)
```
