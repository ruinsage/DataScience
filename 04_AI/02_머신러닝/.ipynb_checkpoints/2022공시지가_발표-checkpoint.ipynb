```python
target = ground_value2['공시지가']
```

```python
ground_data = pd.concat([ground_value2['시도명'], ground_value2['시군구명'], ground_value2['지목'], ground_value2['면적'],
              ground_value2['용도지역1'], ground_value2['용도지역2'], ground_value2['이용상황'], ground_value2['주위환경'],
              ground_value2['지세명'], ground_value2['형상명'], ground_value2['도로교통'],
              ground_value2['방위'], ground_value2['지리적위치1_2'], ground_value2['지리적위치2_2']], axis = 1)
```

```python
ground_data['용도지역1'].replace('제2종전용주거지역','2', inplace =True)
```

```python
ground_data
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(ground_data, target, random_state=142)
```

```python
poly = PolynomialFeatures(include_bias=False)

poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
```

```python
ss = StandardScaler() # 표준화 작업을 함수로서 지원
ss.fit(train_poly)

train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

```python
lasso = Lasso()
lasso.fit(train_scaled, train_target)
print(f'training_score : {lasso.score(train_scaled, train_target)}')
print(f'test_score : {lasso.score(test_scaled, test_target)}')
```

```python
lasso = Lasso(alpha = 1)
print(f'training_score : {lasso.score(train_scaled, train_target)}')
print(f'test_score : {lasso.score(test_scaled, test_target)}')
```
