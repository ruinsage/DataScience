```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import urllib.request
from bs4 import BeautifulSoup
import re
import urllib.request as ur
import pandas as pd
```

```python
kakao = pd.read_csv('쇼핑end.csv',encoding = 'cp949')
kakao.head()
```

```python
name_list = kakao['콘텐츠명']
name_list
```

```python
all_name = []
for i in name_list:
    all_name.append(i)
len(all_name)
```

```python
base_options = Options()
base_options.add_experimental_option("detach", True)
base_service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=base_service, options=base_options)
# 가상의 크롬브라우저 실행
```

```python
url = "https://map.kakao.com/"

all_page = []

for name in all_name:
    restorant_name = '제주도 ' + name
    driver.get(url)
    time.sleep(1)
    driver.find_element(By.CLASS_NAME, "tf_keyword").send_keys(restorant_name)
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="search.keyword.submit"]').send_keys(Keys.ENTER)
    time.sleep(1)

    # 1페이지
    soup1=BeautifulSoup(driver.page_source,'html.parser')
    tags1 = soup1.find_all('li', {"class" : "PlaceItem"})
    location_company=re.compile('[<]a.*href[=]["](https.*place.*#comment)["]\starget')
    company_page=location_company.findall(str(tags1))
    if len(company_page) >= 1:
        print(company_page[0])
        all_page.append(company_page[0])
    else:
        company_page = ['정보없음']
        print(company_page[0])
        all_page.append(company_page[0])
```
