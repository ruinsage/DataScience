# 후기 데이터 전처리

```python
df['CONTENT'] = df['CONTENT'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣|\\n ]","")
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0,subset=['CONTENT'],how='any',inplace = True)
```

```python
df.isnull().sum()
```

```python
len_content = []
for i in df['CONTENT']:
    len_content.append(len(i))
len(len_content)
```

```python
df['len_content'] = len_content
df.head()
```

```python
df = df.sort_values(by=['ID','SCORE','len_content'] ,ascending=False)
df.head()
```

```python
df = df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
df['ID'].value_counts(ascending = False)
```

```python
review_count_mean = df['ID'].value_counts().mean()
round(review_count_mean)
```

```python
flag_first = True
for id in df['ID'].unique():
    if flag_first == True:
        df4 = df.loc[df['ID'] == id].iloc[:round(review_count_mean),:]
        flag_first = False
    else:
        df4 = pd.concat([df4,df.loc[df['ID'] == id].iloc[:round(review_count_mean),:]],axis = 0)
df4.head()
```

```python
len(df4['ID'].unique())
```

```python
df4 = df4.sort_values(by=["ID"], ascending=[True]) 
df4 = df4.drop(['WRITER','TYPE','SCORE','CREATED_AT','len_content'],axis=1)
df4
```

# 후기데이터 문자열 파싱

```python
all_data = []

for id in df4['ID'].unique():
    row = []
    row.append(id)
    content_list = ''
    flag_first = True
    for content in df4['CONTENT'].loc[df4['ID'] == id]:
        if flag_first == True:
            if len(content) > 1: 
                content_list = content
                flag_first = False
        else:
            if len(content) > 1: 
                content_list = content_list + '\n' + content
    row.append(content_list)
    all_data.append(row)
all_data
```

---

# 각 업체 후기 최빈 단어 검색

```python
all_keys = []
okt = Okt()
index_list = []
for index in range(len(all_data)):
    index_list.append(all_data[index][0])
    lines = all_data[index][1].split("\n")
    word_dic = {}
    for line in lines:
        malist = okt.pos(line,norm=True, stem=True)
        for word in malist:
            if word[1] == "Noun" : #  명사 확인하기 --- (※3)
                if len(word[0]) > 1:
                    if not (word[0] in word_dic):
                        word_dic[word[0]] = 0
                    word_dic[word[0]] += 1 # 카운트하기

    # 많이 사용된 명사 출력하기 --- (※4)
    keys = sorted(word_dic.items(), key=lambda x:x[1], reverse=True)
    all_keys.append(keys)
```

```python
all_keys
```

```python
with open("all_keys.pickle","wb") as fw:
    pickle.dump(all_keys, fw)

!dir all_keys.pickle
```

# 해당 검색어가 가장 많은 업체와 유사한 업체 

```python
okt = Okt()
results = []
rest_id_list = []
for id,text in enumerate(all_data):
    lines = text[1].split("\n")
    rest_id_list.append(text[0])
    rl = ''
    r = ''
    for line in lines:
        # 형태소 분석하기 --- (※3)
        # 단어의 기본형 사용
        malist = okt.pos(line, norm=True, stem=True)
        word_list = ''
        for word in malist:
            # 어미/조사/구두점 등은 대상에서 제외 
            if not word[1] in ["Josa", "Eomi", "Punctuation"]:
                if len(word[0]) > 1 :
                    word_list = word_list + ' ' + word[0]
        r = r + word_list
    rl = rl + r
    results.append(rl)
results
```

```python
with open("id_list.pickle","wb") as fw:
    pickle.dump(rest_id_list, fw)

!dir id_list.pickle
```

# 유사업체 찾기 실행문

```python
stopwords = ['이다','이네','이라고','이나','인가','이고','싶다','으로','인데','에는',
             '하다','하고','에서','비다','알다','라고','달다','다시다','지만','에서는']
```

```python
vectorizer = TfidfVectorizer(min_df=2, stop_words= stopwords, ngram_range = (1, 2))
```

```python
X = vectorizer.fit_transform(results)
feature_names = vectorizer.get_feature_names_out()
feature_names
```

```python
rest_sim = cosine_similarity(X)
```
