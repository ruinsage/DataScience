# 각 업체 후기 최빈 단어 검색

```python
all_keys = []
okt = Okt()
index_list = []
for index in range(len(all_data)):
    index_list.append(all_data[index][0])
    lines = all_data[index][1].split("\n")
    word_dic = {}
#     word_dic[word[0]] = 0
    for line in lines:
        malist = okt.pos(line,norm=True, stem=True)
        for word in malist:
            if word[1] == "Noun" : #  명사 확인하기 --- (※3)
                if len(word[0]) > 1:
                    if not (word[0] in word_dic):
                        word_dic[word[0]] = 0
                    word_dic[word[0]] += 1 # 카운트하기

    # 많이 사용된 명사 출력하기 --- (※4)
    keys = sorted(word_dic.items(), key=lambda x:x[1], reverse=True)
    all_keys.append(keys)
```

# 해당 검색어가 가장 많은 업체와 유사한 업체 

```python
okt = Okt()
results = []
rest_id_list = []
for id,text in enumerate(all_data):
    lines = text[1].split("\n")
    rest_id_list.append(text[0])
    rl = ''
    r = ''
    for line in lines:
        # 형태소 분석하기 --- (※3)
        # 단어의 기본형 사용
        malist = okt.pos(line, norm=True, stem=True)

        word_list = ''
        for word in malist:
            # 어미/조사/구두점 등은 대상에서 제외 
            if not word[1] in ["Josa", "Eomi", "Punctuation"]:
                if len(word[0]) > 1 :
                    word_list = word_list + ' ' + word[0]
        r = r + word_list
    rl = rl + r
    results.append(rl)
results
```

# 유사업체 찾기 실행문

```python
stopwords = ['이다','이네','이라고','이나','인가','이고','싶다','으로','인데','에는',
             '하다','하고','에서','비다','알다','라고','달다','다시다','지만','에서는']
```

```python
vectorizer = TfidfVectorizer(min_df=2, stop_words= stopwords)
```

```python
X = vectorizer.fit_transform(results)
feature_names = vectorizer.get_feature_names_out()
feature_names
```

```python
len(feature_names)
```

```python
rest_sim = cosine_similarity(X)
rest_sim[:3]
```

```python
def similar_recommend_by_rest_id(rest_id, n_recommand):
    rest_index = rest_id
    recommended_store = []
    similar_rest = sorted(list(enumerate(rest_sim[rest_index])),key=lambda x:x[1], reverse=True)
    recommended=1

    for rest_info in similar_rest[0:n_recommand]:
        rest_title= rest_id_list[rest_info[0]]
        recommended_store.append(rest_title)
    return(recommended_store)
```
