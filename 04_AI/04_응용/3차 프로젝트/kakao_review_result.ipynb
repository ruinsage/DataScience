### 후기정보 데이터베이스 작성

```python
df = data_frame.iloc[:,1:3]
df.head()
```

```python
df['콘텐츠명'].unique()
```

```python
df_kakao['STORE_NAME'].unique()
```

```python
df.shape
```

```python
df_kakao.shape
```

```python
df2 = pd.merge(df, df_kakao, how='inner', left_on = '콘텐츠명', right_on = 'STORE_NAME')
df2.head()
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
df3 = df2.drop(['콘텐츠명','OPEN_CLOSE_TIME','CATEGORY','STORE_NAME','STORE_ADDR'],axis=1)
df3
```

```python
df3.shape
```

```python
df4 = df3.copy()
```

```python
df4.rename(columns={'콘텐츠 ID':'ID'},inplace=True)
```

```python
df4.head()
```

```python
df4.fillna(' ',inplace=True)
```

```python
df4.info()
```

```python
df4.isnull().sum()
```

```python
df4.shape
```

```python
df4.rename(columns={'STORE_SCORE':'SCORE'},inplace = True)
```

```python
df4['CONTENT']=df4['CONTENT'].str.replace('<br/>','')
df4['SCORE'] = df4['SCORE'].replace('별점\n', '', regex=True)
df4['SCORE'] = df4['SCORE'].replace('/5', '', regex=True)
df4['TYPE']=df4['TYPE'].str.replace('네이버','naver')
```

```python
df4.shape
```

```python
df4 = df4[['ID','WRITER','CONTENT','TYPE','SCORE','CREATED_AT']]
df4
```

### 업체정보 데이터 베이스 파일 작성

```python
data_frame.head(1)
```

```python
data_frame2 = data_frame.drop(['관련 태그 전체','주소','태그','우편번호'],axis=1)
data_frame2
```

```python
df2.head()
```

```python
df2_pre = df2.drop(['콘텐츠명','WRITER','CONTENT','TYPE','SCORE','CREATED_AT',
                    'STORE_NAME','STORE_ADDR','ECT'],axis=1)
df2_pre
```

```python
df2_pre.shape
```

```python
df2_pre = df2_pre.drop_duplicates(['콘텐츠 ID'], keep='first')
df2_pre.shape
```

```python
data_frame2.shape
```

```python
data_frame3 = pd.merge(data_frame2, df2_pre, how='left', left_on = '콘텐츠 ID', right_on = '콘텐츠 ID')
data_frame3.head()
```

```python
data_frame3.shape
```

```python
data_frame3.rename(columns={'콘텐츠 ID':'ID','콘텐츠명':'TITLE','간단소개':'DESCRIPTION',
                            '도로명주소':'ADDRESS','위도':'LAT','경도':'LON','전화번호':'TEL',
                            'OPEN_CLOSE_TIME':'OPEN_TIME','HOMPAGE':'PAGE'},inplace=True)
```

```python
data_frame3.rename(columns={'STORE_SCORE':'STORE_OF_NAVER'},inplace=True)
# 필요한경우 STORE_OF_KAKAO 변경
```
