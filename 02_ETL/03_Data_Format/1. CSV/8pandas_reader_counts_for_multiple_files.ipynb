```python
import pandas as pd
```

```python
import glob
```

```python
import os
```

```python
input_path = '.'
```

```python
file_list = glob.glob(os.path.join(input_path,'sales_*'))
file_list
```

```python
sales_february_2014_df = pd.read_csv(file_list[0])
sales_february_2014_df
```

```python
sales_january_2014_df = pd.read_csv(file_list[1])
sales_january_2014_df
```

```python
sales_march_2014_df = pd.read_csv(file_list[2])
sales_march_2014_df
```

```python
result = pd.concat([sales_february_2014_df,sales_january_2014_df,sales_march_2014_df]).reset_index(drop=True)
result
```

```python
result.to_csv('9pandas_output_basic', index = False)
```
