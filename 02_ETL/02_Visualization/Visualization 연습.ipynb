## 3. numpy.random.randint

[이산균등분포(discrete uniform)](https://ko.wikipedia.org/wiki/%EC%9D%B4%EC%82%B0%EA%B7%A0%EB%93%B1%EB%B6%84%ED%8F%AC)

1) low 가장 작은 범주가 어디서부터 시작하는 지를 적습니다. 다만 1개의 hyperparameter를 적으면 그 값은 가장 높은 범주의 값으로 인식하고 0에서부터 그 숫자 사이의 int를 1개 반환합니다. 만약 high없이 개수를 늘리고 싶다면 다음과 같이 적으면 됩니다.
ex) numpy.random.randint(3,size = 10)

2) high 가장 높은 정수를 적는 곳으로 선택사항입니다.

3) size 몇개의 random 값을 리턴하려는지 적으면 됩니다. 3개 반환 받고 싶다면 3
   만약 array형태로 shape를 적고 싶다면 다음과 같이 적을 수도 있습니다.
ex) numpy.random.randint(3,10,size = (3,2))

4) dtype int를 반환하는 함수이기 때문에 굳이 필요한가 싶지만 이 부분은 int 혹은 int64를 반환할 수 있도록 선택하는 parameter입니다. 선택하지 않아도 무방합니다.

```python
move_right =  172.5
variance_amplication = 5
```

```python
height_data = np.round(
    move_right +
    (np.random.randn(DATA_LIMIT)*variance_amplication),
    1)
```

```python
height_data
```

```python
plt.xlabel('키')
plt.ylabel('누적개수')
plt.hist(height_data,bins = 140)
```

```python
m_move_right =  172.5
w_move_right =  159.6
variance_amplication = 7
m_height_data = np.round(m_move_right + (np.random.randn(DATA_LIMIT)*variance_amplication),1)
w_height_data = np.round(w_move_right + (np.random.randn(DATA_LIMIT)*variance_amplication),1)

plt.xlabel('키')
plt.ylabel('누적개수')
plt.hist((m_height_data,w_height_data), bins=200, label=['성인 남자키','성인 여자키'])

plt.title('전국 성인남녀 키 현황 (2022.3 산업통산자원부 통계자료 시뮬레이션)')
plt.legend()

plt.show()
```
