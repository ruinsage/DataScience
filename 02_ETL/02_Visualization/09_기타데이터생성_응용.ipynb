```python

from numpy.random import randn
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# cumsum() 값을 누적
plot_data1 = randn(50).cumsum()
plot_data2 = randn(50).cumsum()
plot_data3 = randn(50).cumsum()
plot_data4 = randn(50).cumsum()

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.plot(plot_data1, marker=r'o', color=u'blue', linestyle='-', label='Blue Solid')
ax1.plot(plot_data2, marker=r'+', color=u'red', linestyle='--', label='Red Dashed')
ax1.plot(plot_data3, marker=r'*', color=u'green', linestyle='-.', label='Green Dash Dot')
ax1.plot(plot_data4, marker=r's', color=u'orange', linestyle=':', label='Orange Dotted')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

ax1.set_title('Line Plots: Markers, Colors, and Linestyles')
plt.xlabel('Draw')
plt.ylabel('Random Number')
# plt.legend()
# loc='best':  그래프 상에서 범례(legend)를 가장 적절한 위치에 배치
plt.legend(loc='best')

plt.show()
```

```python
from numpy.random import randn
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# cumsum() 값을 누적
DATA_LIMIT = 25
plot_data1 = (randn(DATA_LIMIT)/10).cumsum()
plot_data2 = (randn(DATA_LIMIT)/10).cumsum()
plot_data3 = (randn(DATA_LIMIT)/10).cumsum()
plot_data4 = (randn(DATA_LIMIT)/10).cumsum()

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.plot(plot_data1, marker=r'o', color=u'blue', linestyle='-', label='백종원')
ax1.plot(plot_data2, marker=r'+', color=u'red', linestyle='--', label='빌게이츠')
ax1.plot(plot_data3, marker=r'*', color=u'green', linestyle='-.', label='이재용')
ax1.plot(plot_data4, marker=r's', color=u'orange', linestyle=':', label='워렌버핏')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

ax1.set_title('제1회 비트코인 트레이딩 수익률 대회')
plt.xlabel('시간')
plt.ylabel('수익률')
# plt.legend()
# loc='best':  그래프 상에서 범례(legend)를 최상의 위치에 배치
plt.xticks(range(DATA_LIMIT))
plt.legend(loc='best')

plt.show()
```

```python
x = np.arange(start=1, stop=15, step=1)
# 선형적인 데이터에 결측치가 생겼을 경우나
# 미래예측 데이터를 구하는 경우
# 비교적 합리적으로 결측치를 채우기위해 
# 데이터의 경향을 파악해 적절한 패턴(공식)을 세우자

# 1차 방정식에 준하는 선형회귀 결과를 얻기 위한 공식
y_linear = x + 5 * np.random.randn(14)
# 2차 방정식에 준하는 선형회귀 결과를 얻기 위한 공식
y_quadratic = x**2 + 10 * np.random.randn(14)

# deg=1 : 1차 방정식 형태, deg=2 : 2차 방정식 형태
fn_linear = np.poly1d(np.polyfit(x, y_linear, deg=1))
fn_quadratic = np.poly1d(np.polyfit(x, y_quadratic, deg=2))

plt.plot(x, y_linear, 'bo', x, y_quadratic, 'go', \
			x, fn_linear(x), 'b-', x, fn_quadratic(x), 'g-', linewidth=2.)


plt.title('1차/2차 방정식 데이터 시뮬레이션')
plt.xlabel('x')
plt.ylabel('f(x)')

# x축, y축 표시 영역 범위 설정
plt.xlim((min(x)-1, max(x)+1))
plt.ylim((min(y_quadratic)-10, max(y_quadratic)+10))

plt.show()
```

```python
DATA_LIMIT = 36

x = np.arange(start=1, stop=DATA_LIMIT+1, step=1)

# 1차 방정식에 준하는 선형회귀 결과를 얻기 위한 공식
y_linear = x + 5 * np.random.randn(DATA_LIMIT)
# 2차 방정식에 준하는 선형회귀 결과를 얻기 위한 공식
y_quadratic = x**2 + 10 * np.random.randn(DATA_LIMIT)

# deg=1 : 1차 방정식 형태, deg=2 : 2차 방정식 형태
fn_linear = np.poly1d(np.polyfit(x, y_linear, deg=1))
fn_quadratic = np.poly1d(np.polyfit(x, y_quadratic, deg=2))

fig = plt.figure()

plt.plot(x, y_linear, 'bo', linewidth=2, label='A사 신약')
plt.plot(x, fn_linear(x), 'b-',  linewidth=2, label='A사 신약 지표 추세')
plt.plot(x, y_quadratic, 'g^', linewidth=2. ,label='B사 신약')
plt.plot(x, fn_quadratic(x), 'g-', linewidth=2. ,label='B사 신약 지표 추세')


plt.title('36개월간 A/B사 신약 투입 임상 테스트 결과')
plt.xlabel('개월수')
plt.ylabel('혈당수치 개선 지표')

# x축, y축 표시 영역 범위 설정
plt.xlim((min(x)-1, max(x)+1))
plt.ylim((min(y_quadratic)-10, max(y_quadratic)+10))

plt.legend()
plt.show()
```

```python
DATA_LIMIT = 36

x = np.arange(start=1, stop=DATA_LIMIT+1, step=1)

# 1차 방정식에 준하는 선형회귀 결과를 얻기 위한 공식
y_linear = x + 5 * np.random.randn(DATA_LIMIT)
# 2차 방정식에 준하는 선형회귀 결과를 얻기 위한 공식
y_quadratic = x**2 + 10 * np.random.randn(DATA_LIMIT)

# np.polyfit(x, y, n)함수는 y의 데이터에대한 최적의 피팅(최소제곱 관점에서)인
# n차 다항식 p(x)의 계수를 반환한다. p의 계수는 내림차순으로 나열되고
# p의 길이는 n+1이다.

# np.poly1d(입력배열,근 설정, 변수명)함수는 입력배열을 다항식의 계수나 근으로
# 사용하여 다항식을 생성한다. 근설정이 True라면 근으로 사용하며 default는 False이다.
# 변수명은 다항식의 이름으로 사용하며 default는 x이다.

# deg=1 : 1차 방정식 형태, deg=2 : 2차 방정식 형태
fn_linear = np.poly1d(np.polyfit(x, y_linear, deg=1))
fn_quadratic = np.poly1d(np.polyfit(x, y_quadratic, deg=2))

fig = plt.figure()

plt.plot(x, y_linear, 'bo', linewidth=2, label='A사 신약')
plt.plot(x, fn_linear(x), 'b-',  linewidth=2, label='A사 신약 지표 추세')
plt.plot(x, y_quadratic, 'g^', linewidth=2. ,label='B사 신약')
plt.plot(x, fn_quadratic(x), 'g-', linewidth=2. ,label='B사 신약 지표 추세')

predict_x = DATA_LIMIT+1
predict_y = fn_quadratic(predict_x)
plt.scatter(predict_x, predict_y)

plt.title('36개월간 A/B사 신약 투입 임상 테스트 결과\n및 37개월차 B사 신약 성능 예측')
plt.xlabel('개월수')
plt.ylabel('혈당수치 개선 지표')

# x축, y축 표시 영역 범위 설정
plt.xlim((min(x)-1, predict_x+2))
plt.ylim((min(y_quadratic)-10, predict_y+100))
plt.text(predict_x-18,predict_y-19,f'37개월차 B사 신약 예상지표 ({round(predict_y)})', c='red')

plt.legend()
plt.show()
```
