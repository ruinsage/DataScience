```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager,rc

import random

font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 박스 플롯은 완벽한 정규분포를 표현하지는 않는다. 대략적으로만 추측할 수 있다.
# 바이올린 플롯을 통해 실제 데이터 분포를 확인할 수 있다.
df = sns.load_dataset("titanic")
sns.violinplot(x=df["age"])
# seaborn의 바이올린 플롯은 안쪽에 수염,1사분위, 중위수, 4사분위등을 표현해준다. (마치 박스 플롯처럼)
```

```python
sns.violinplot(data=df, x="age", y="class")
# y축을 지정하면 그룹바이한 것처럼 표현된다.
```

```python
sns.violinplot(data=df, x="class", y="age", hue="alive")
```

```python
# inner 인자에 아래 값으로 설정 가능
# {“box”, “quartile”, “point”, “stick”, None}
# stick의 밀집도에따라 데이터의 분포, 밀도 반영
sns.violinplot(data=df, x="age", y="embark_town", inner="stick")
```

```python
sns.violinplot(data=df, x="age", y="embark_town", inner="point")
# 점의 밀집도에따라 데이터의 밀도를 반영
```

```python
sns.violinplot(data=df, x="age", y="embark_town", inner="quartile")
```

```python
sns.violinplot(data=df, x="age", y="embark_town", hue = 'sex' ,inner= None)
```
