# Pairplot : 수치형 데이터의 모든 상관관계를 분석하기 위한 차트
### 데이터프레임을 pairplot의 인자에 넣어주면 수치형 데이터의 모든 상관관계를 산점도 형식으로 보여준다

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# 좌상단부터 우하단까지 대각선상에 있는 셀은 자기 자신이다.
# 자기 자신은 상관관계를 표현할 수 없으므로 히스토그램 정보만 보여준다.
sns.pairplot(penguins)
plt.show()
```

```python
sns.pairplot(penguins, hue="species")
plt.show()
```

```python
sns.pairplot(penguins, hue="species", diag_kind="hist")
plt.show()
```

```python
sns.pairplot(penguins, hue="species", markers=["o", "s", "D"])
plt.show()
```

```python
sns.pairplot(penguins, hue="species", markers=["o", "s", "D"], corner=True)
plt.show()
# corner=True 옵션은 중복되는 차트는 생략한다.
# 유니크한 조합의 pairplot만 보고자 할 때 유용하다.
```
