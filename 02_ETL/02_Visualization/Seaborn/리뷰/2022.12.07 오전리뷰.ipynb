```python
# 박스 플롯은 완벽한 정규분포를 표현하지는 않는다. 대략적으로만 추측할 수 있다.
# 바이올린 플롯을 통해 실제 데이터 분포를 확인할 수 있다.
sns.violinplot(x=df["age"])
# seaborn의 바이올린 플롯은 안쪽에 수염,1사분위, 중위수, 4사분위등을 표현해준다.
# (마치 박스 플롯처럼)
```

```python
# inner 인자에 아래 값으로 설정 가능하다.
# {“box”, “quartile”, “point”, “stick”, None}
# inner="stick" 옵션은 선의 밀집도에따라 데이터의 분포, 밀도 반영한다.
# inner="point" 옵션은 점의 밀집도에따라 데이터의 밀도를 반영한다.
# inner="box" 옵션은 박스 플롯처럼 내부에 데이터선을 표현한다.
# inner= None 옵션은 오직 바이올린 플롯만 표현한다.
sns.violinplot(data=df, x="age", y="embark_town", hue = 'sex' ,inner="box")
```

```python
df = sns.load_dataset("glue")
df.shape
```

```python
glue = df.pivot("Model", "Task", "Score")
glue.shape
```

```python
glue.head(8)
```

```python
# annot옵션에서 데이터프레임.랭크함수를 호출하면 axis 방향에따라
# ascending 순서로 순위를 메긴다.
# linewidth옵션은 칸마다 간격을 조절할 수 있다.
sns.heatmap(glue, annot=glue.rank(axis="columns", ascending=False), linewidth=1)
```

```python
# annot True로 설정하게 되면 각 히트맵 셀의 값을 표시해준다.
# fmt욥션의 값으로 실수 표현 스트링을 정의
# cmap옵션으로 다른 컬러맵을 적용할 수 있다.
# 디폴트 컬러맵은 높은 값일 수록 진하게 또는 강하게 표현되는 색깔톤과 어울리지 않는다.
sns.heatmap(glue, annot=glue.rank(), linewidth=1, cmap='crest', fmt=".1f")
```

```python
# kind에 kde 옵션을 주면 변량과 밀도를 반영하여 등고선 형태로 데이터 분포를 표시
# kind에 hist 옵션을 주면 사각형 모자이크 형태로 변량과 밀도는 색깔로 반영하여 데이터 분포를 표시
# kind에 hex 옵션을 주면 육각형 벌집 형태로 변량과 밀도는 색깔을 반영하여 데이터 분포를 표시
# kind에 reg 옵션을 주면 데이터의 linear regression 기반 추세선을 표시
# 만약 데이터가 선형 관게가 아니라면 추세선의 정확도, 신뢰도가 떨어진다.
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind="reg")
```

```python
# 좌상단부터 우하단까지 대각선상에 있는 셀은 자기 자신이다.
# 자기 자신은 상관관계를 표현할 수 없으므로 히스토그램 정보만 보여준다.
sns.pairplot(penguins, hue="species", diag_kind="hist", markers=["o", "s", "D"], corner=True)
plt.show()
# corner=True 옵션은 중복되는 차트는 생략한다.
# 유니크한 조합의 pairplot만 보고자 할 때 유용하다.
```
