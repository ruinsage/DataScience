```python
df = sns.load_dataset("glue")
df.shape
```

```python
glue = df.pivot("Model", "Task", "Score")
glue.shape
```

```python
glue.head(8)
```

```python
# annotì˜µì…˜ì—ì„œ ë°ì´í„°í”„ë ˆìž„.ëž­í¬í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ë©´ axis ë°©í–¥ì—ë”°ë¼
# ascending ìˆœì„œë¡œ ìˆœìœ„ë¥¼ ë©”ê¸´ë‹¤.
# linewidthì˜µì…˜ì€ ì¹¸ë§ˆë‹¤ ê°„ê²©ì„ ì¡°ì ˆí•  ìˆ˜ ìžˆë‹¤.
sns.heatmap(glue, annot=glue.rank(axis="columns", ascending=False), linewidth=1)
```

```python
# annot Trueë¡œ ì„¤ì •í•˜ê²Œ ë˜ë©´ ê° ížˆíŠ¸ë§µ ì…€ì˜ ê°’ì„ í‘œì‹œí•´ì¤€ë‹¤.
# fmtìš¥ì…˜ì˜ ê°’ìœ¼ë¡œ ì‹¤ìˆ˜ í‘œí˜„ ìŠ¤íŠ¸ë§ì„ ì •ì˜
# cmapì˜µì…˜ìœ¼ë¡œ ë‹¤ë¥¸ ì»¬ëŸ¬ë§µì„ ì ìš©í•  ìˆ˜ ìžˆë‹¤.
# ë””í´íŠ¸ ì»¬ëŸ¬ë§µì€ ë†’ì€ ê°’ì¼ ìˆ˜ë¡ ì§„í•˜ê²Œ ë˜ëŠ” ê°•í•˜ê²Œ í‘œí˜„ë˜ëŠ” ìƒ‰ê¹”í†¤ê³¼ ì–´ìš¸ë¦¬ì§€ ì•ŠëŠ”ë‹¤.
sns.heatmap(glue, annot=glue.rank(), linewidth=1, cmap='crest', fmt=".1f")
```

```python
# kindì— kde ì˜µì…˜ì„ ì£¼ë©´ ë³€ëŸ‰ê³¼ ë°€ë„ë¥¼ ë°˜ì˜í•˜ì—¬ ë“±ê³ ì„  í˜•íƒœë¡œ ë°ì´í„° ë¶„í¬ë¥¼ í‘œì‹œ
# kindì— hist ì˜µì…˜ì„ ì£¼ë©´ ì‚¬ê°í˜• ëª¨ìžì´í¬ í˜•íƒœë¡œ ë³€ëŸ‰ê³¼ ë°€ë„ëŠ” ìƒ‰ê¹”ë¡œ ë°˜ì˜í•˜ì—¬ ë°ì´í„° ë¶„í¬ë¥¼ í‘œì‹œ
# kindì— hex ì˜µì…˜ì„ ì£¼ë©´ ìœ¡ê°í˜• ë²Œì§‘ í˜•íƒœë¡œ ë³€ëŸ‰ê³¼ ë°€ë„ëŠ” ìƒ‰ê¹”ì„ ë°˜ì˜í•˜ì—¬ ë°ì´í„° ë¶„í¬ë¥¼ í‘œì‹œ
# kindì— reg ì˜µì…˜ì„ ì£¼ë©´ ë°ì´í„°ì˜ linear regression ê¸°ë°˜ ì¶”ì„¸ì„ ì„ í‘œì‹œ
# ë§Œì•½ ë°ì´í„°ê°€ ì„ í˜• ê´€ê²Œê°€ ì•„ë‹ˆë¼ë©´ ì¶”ì„¸ì„ ì˜ ì •í™•ë„, ì‹ ë¢°ë„ê°€ ë–¨ì–´ì§„ë‹¤.
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
```
