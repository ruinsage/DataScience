```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
from numpy.random import randn
import matplotlib.pyplot as plt
import random

font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
DATA_LIMIT = 100

horse1 = (abs(randn(DATA_LIMIT)+randn(1)+randn(1))*10).cumsum()
horse2 = (abs(randn(DATA_LIMIT)+randn(1)+randn(1))*10).cumsum()
horse3 = (abs(randn(DATA_LIMIT)+randn(1)+randn(1))*10).cumsum()
horse4 = (abs(randn(DATA_LIMIT)+randn(1)+randn(1))*10).cumsum()
horse5 = (abs(randn(DATA_LIMIT)+randn(1)*randn(1))*10).cumsum()
horse6 = (abs(randn(DATA_LIMIT)+randn(1)*randn(1))*10).cumsum()
horse7 = (abs(randn(DATA_LIMIT)+randn(1)*randn(1))*10).cumsum()
horse8 = (abs(randn(DATA_LIMIT)+randn(1)*randn(1))*10).cumsum()

horse1_record = []
horse2_record = []
horse3_record = []
horse4_record = []
horse5_record = []
horse6_record = []
horse7_record = []
horse8_record = []

is_break = False
for index in range(DATA_LIMIT):
    horse1_record.append(horse1[index])
    horse2_record.append(horse2[index])
    horse3_record.append(horse3[index])    
    horse4_record.append(horse4[index])
    horse5_record.append(horse5[index])
    horse6_record.append(horse6[index])
    horse7_record.append(horse7[index])
    horse8_record.append(horse8[index])
    
    if is_break == True:
        break
    else:
        pass
    if horse1[index] > 1000:
        print(f'라이스샤워 우승 기록: {index}초')
        is_break = True
    if horse2[index] > 1000:
        print(f'티엠오페라오 우승 기록: {index}초')
        is_break = True
    if horse3[index] > 1000:
        print(f'보드카 우승 기록: {index}초')
        is_break = True
    if horse4[index] > 1000:
        print(f'에어그루브 우승 기록: {index}초')
        is_break = True
    if horse5[index] > 1000:
        print(f'엘콘돌파사 우승 기록: {index}초')
        is_break = True
    if horse6[index] > 1000:
        print(f'스페셜위크 우승 기록: {index}초')
        is_break = True
    if horse7[index] > 1000:
        print(f'골드쉽 우승 기록: {index}초')
        is_break = True
    if horse8[index] > 1000:
        print(f'위닝티켓 우승 기록: {index}초')
        is_break = True
    
fig = plt.figure(figsize = (10,3))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(horse1_record, color=u'black', linestyle=':', label='라이스샤워')
ax1.plot(horse2_record, color=u'orange', linestyle=':', label='티엠오페라오')
ax1.plot(horse3_record, color=u'brown', linestyle=':', label='보드카')
ax1.plot(horse4_record, color=u'blue', linestyle=':', label='에어그루브')
ax1.plot(horse5_record, color=u'red', linestyle=':', label='엘콘돌파사')
ax1.plot(horse6_record, color=u'purple', linestyle=':', label='스페셜위크')
ax1.plot(horse7_record, color=u'gold', linestyle=':', label='골드쉽')
ax1.plot(horse8_record, color=u'green', linestyle=':', label='위닝티켓')

ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
plt.axhline(1000,0,1, color = 'lightgray', linestyle = '--', linewidth = 1)

ax1.text(18, 990, '결승선 라인')
ax1.set_title('GRADE 1 경마 대회')
plt.xlabel('시간(초)')
plt.ylabel('거리(미터)')

plt.legend(loc='best')
plt.show()
```
