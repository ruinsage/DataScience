```python
plt.style.use('Solarize_Light2')
plt.rcParams['figure.figsize'] = (8, 4)
plt.rcParams['font.size'] = 14

# fig는 피규어 타입 객체
# 즉, ax1변수를 피규어타입으로 저장
fig, ax = plt.subplots()

ax.set_title('Mean Squared Error', pad=20)
ax.set_xlim(-4, 4)
ax.set_ylim(0, 5)
ax.set_xticks([-3, -2, -1, 0, 1, 2, 3])
ax.set_yticks([1, 2, 3, 4, 5])

x = np.linspace(-3, 3, 100)
ax.set_xlabel('y_true - y_pred', fontdict={'fontsize': 14}, labelpad=10)
ax.plot(x, x**2, color='#4799FF', linewidth=5)

ax.spines['left'].set_position('center')        # 왼쪽 축을 가운데 위치로 이동
ax.spines['right'].set_visible(False)          # 오른쪽 축을 보이지 않도록
ax.spines['top'].set_visible(False)            # 위 축을 보이지 않도록
ax.tick_params('both', length=0)                # x,y축 Tick의 눈금 길이 0

plt.show()
plt.savefig('style_spines_options.png', dpi = 50)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 14

x = np.arange(0, 3)
y1 = x + 1
y2 = [6,3,2]
# fig는 피규어 타입 객체
# 즉, ax1변수를 피규어타입으로 저장
fig, ax1 = plt.subplots()
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('1st Y-Axis')
line1 = ax1.plot(x, y1, color='green', label='1st Data')

ax2 = ax1.twinx()
ax2.set_ylabel('2nd Y-Axis')
line2 = ax2.plot(x, y2, color='deeppink', label='2nd Data')

lines = line1 + line2
# for문 사용 (복잡한 차트의 경우 자동화 할 수 있다)
labels = [l.get_label() for l in lines]
# 변수명 = i.라벨을 꺼내다()반복 lines에서
# fo문 결과 labels = ['1st Data', '2nd Data']
ax1.legend(lines, labels, loc=(1.2, 0.0))
# 범례 위치 변경
plt.show()
plt.savefig('double_line.png', dpi = 200)
```

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 2.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()
# vert = > 수평표시 옵션
# whis 옵션 : float 숫자로 수염의 범위를 설정한다. 숫자가 높을수록 
# 수염의 범위가 넓어져 fliers값이 줄어든다.
box = ax.boxplot([data_a, data_b, data_c], notch=True, whis=1.5)
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
# for문 사용 (복잡한 차트의 경우 자동화 할 수 있다)
whiskers = [item.get_ydata() for item in box['whiskers']]
medians = [item.get_ydata() for item in box['medians']]
fliers = [item.get_ydata() for item in box['fliers']]
# 변수명 = i.라벨을 꺼내다()반복 lines에서
print('whiskers:', whiskers)
print('medians:', medians)

# filiers: 수염 범위 밖의 데이터(Outliers), 이상점 + 극단점
print('fliers:', fliers)

plt.show()
plt.savefig('double_line.png', dpi = 400)
```

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 2.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True, quantiles=[[0.25, 0.75], [0.25, 0.75], [0.25, 0.75]])
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

violin['cbars'].set_edgecolor('green')
violin['cmaxes'].set_edgecolor('red')
violin['cmins'].set_edgecolor('blue')
violin['cmeans'].set_edgecolor('gray')

# 도형을 여러개 넣을 수는 없는듯?
# 4. 다각형 그리기
ax.add_patch(
     patches.Polygon(
        ((1.9, 4.0), (1.9, 2.5), (2.1,2.5), (2.1, 4.0),(2.0,5.0)), # (x, y)
        closed=True,                                              # width, height
        edgecolor = 'deeppink',
        facecolor = 'lightgray'
     ))

# 5. 텍스트 삽입하기
ax.text(1.5, 5.0, 'Max of Data B')

plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 5)
plt.rcParams['font.size'] = 12

x = [1, 2, 3]
y = [1, 2, 3]

fig, ((ax1, ax2), (ax3, ax4),(ax5, ax6)) = plt.subplots(3, 2)

ax1.bar(x, y, color='aquamarine', edgecolor='black', hatch='/')
ax2.bar(x, y, color='salmon', edgecolor='black', hatch='\\')
ax3.bar(x, y, color='navajowhite', edgecolor='black', hatch='+')
ax4.bar(x, y, color='lightskyblue', edgecolor='black', hatch='*')
ax5.bar(x, y, color='lightskyblue', edgecolor='black', hatch='|')
ax6.bar(x, y, color='lightskyblue', edgecolor='black', hatch='-')

plt.tight_layout()
plt.show()
```

```python

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = [1, 2, 3]
y = [1, 2, 3]

fig, ax = plt.subplots()
# 각 bar를 따로 무늬 지정
bars = ax.bar(x, y, color='lightgray', edgecolor='black')
bars[0].set_hatch('x')
bars[1].set_hatch('O')
bars[2].set_hatch('.')

plt.tight_layout()
plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12

x = [1,2,3]
y = [1,2,3]

for i in range(6):
    hatch_str = "/" * i

    ax = plt.subplot(2, 3, i + 1)
    ax.set_title("Hatch String: " + hatch_str)
    bars = ax.bar(x,y,facecolor='skyblue', edgecolor='black')

    for bar in bars:
# hatch 문자열의 개수를 한 개 이상 입력함으로써 패턴의 밀도를 지정
        bar.set_hatch(hatch_str)

plt.tight_layout()
plt.show()
```
