## 파이차트 그리기

```python
MAX_RECORD = 1541962
treatment_range = ['완치자','격리자','중환자','사망자']
treatment_list = []

for i in range(MAX_RECORD):
    treatment_list.append(treatment_range[random.choices(range(0,len(treatment_range)), weights = [0.6,0.25,0.1,0.05])[0]])
    
```

```python
df2 = pd.DataFrame(treatment_list,
                   columns = ['현재상태'])
```

```python
df2.head()
```

```python
value_count_list = pd.Series(treatment_list).value_counts()
```

```python
value_count_list
```

```python
explode = [0,0,0.1,0.1]
colors = ['#229966','#ff2233','#00ff44','silver']
wedgeprops = {'width' : 0.6, 'edgecolor' : 'w', 'linewidth' : 5}
# 부채꼴스타일 {원넓이: 수치, 모서리색: 색깔, 선넓이 : 수치}

plt.pie(value_count_list, labels = treatment_range, startangle = 270, counterclock = False, explode = explode,
        colors = colors, wedgeprops = wedgeprops, shadow = True, autopct = '%d%%')
plt.show
```

```python
import numpy as np
import random
```

```python
keyword = ['대한민국','손흥민','맥주','이마트','FIFA',
           '배달','승부예측','교촌치킨','배민','안주류']

np.random.seed(2)
n = 10
y = np.random.randint(0,10000, size = n)
```
