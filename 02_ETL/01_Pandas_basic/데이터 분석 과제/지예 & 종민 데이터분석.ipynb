```python
import pandas as pd
```

# 1. book_data라는 csv파일을 가져온다

```python
book_data = pd.read_csv("data/book_data.csv",encoding='euc-kr')
```

```python
book_data
```

```python
book_data.isnull().sum()
```

# 2. book_cover 열의 NaN 값을 "자료없음" 으로 바꾸기

```python
book_data['book_cover'].fillna("자료없음",inplace=True)
book_data[['book_cover']]
```

# 3. 카테고리에서 에세이 인 항목만 검색

```python
book_data.loc[book_data['book_category'] == '에세이'].head(10)
```

---

# 4. 원하는 열만 선택해서 사용

```python
best_saler=book_data[['book_name','book_date','book_publisher', 'sales_point']]
```

```python
best_saler
```

```python
best_saler.dtypes
```

# 5. sales_point 타입의 숫자를 사용하기 위해 data_type 변환

```python
best_saler['sales_point'] =best_saler.loc[:,'sales_point'].str.replace(',','').astype('int')
```

```python
best_saler.dtypes
```

# 6. 판매량이 높은 순서대로 top10 데이터 확보

```python
best_saler = best_saler.nlargest(100,'sales_point')
```

```python
best_saler.head(10)
```

# 7. Top 100 순위 안의 책을 가장 많이 출간한 출판사 도출

```python
best_publisher = best_saler['book_publisher'].value_counts().head()
```
