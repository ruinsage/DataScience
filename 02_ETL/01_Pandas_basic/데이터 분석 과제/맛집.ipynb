# 맛집리스트 

```python
matzip = pd.read_csv('../data/matzipData.csv', encoding='UTF-8')
```

# 칼럼 데이터 타입 변경

```python
matzip.isnull().sum()
```

```python
matzip.replace({'Rating' : '-'}, 0, inplace=True)
```

```python
matzip['Rating'] = matzip['Rating'].astype(float)
```

# 특정조건 데이터 추출

```python
best_matzip = matzip[matzip['Rating'] >=  3.5]
```
