# 예제 7-2 통계함수 사용하기 (기초)

```python
sample_df= pd.DataFrame(
    [
        [9.1,1.2,1.3,1.4],
        [2.1,9.2,9.3,2.4],
        [3.1,3.2,3.3,3.4]
    ],
    index =['a','b','c'], columns=['A','B','C','D']
)
sample_df
```

```python
# describe() 함수는 열별로 간단한 통계결과를 설명하여 보여준다.
# 기본 옵션으로 describe() 함수 수행시 수치형 데이터에대한 통계값을 제공한다.
sample_df.describe()
# std 표준편차
# 25% 1사분위 수 (1st quntile)
# 50% 2사분위 수(중앙값,Median, 2st quntile)
# 75% 3사분위 수 (3rd quntile)
```

```python
sample_df['A'].describe()
```

```python
sample_df.describe(percentiles = [.01,.25,.75,.99])
# percentiles 인자로 사용자가 quntile을 지정할 수 있다
```

```python
sample_df2= pd.DataFrame(
    [
        ['a',1.2,1.3,1.4],
        ['c',9.2,9.3,2.4],
        ['d',3.2,3.3,3.4],
        ['c',4.7,2.3,5.4]
    ],
    index =['a','b','c','d'], columns=['A','B','C','D']
)
sample_df2
```

```python
sample_df2.describe()
```

```python
# 문자열로 구성된 열에 대한 통계값을 보여준다.
sample_df2.describe(include=['object'])
# count: 전체 관측값
# unique: 유일값
# top: 가장 빈도가 높은 데이터
# freq: 가장 빈도가 높은 데이터의 개수 (Top의 개수)
```

```python
HeroInfo = pd.read_csv("../data/HeroInfo.csv",encoding='utf-8')

for i in range(0,748) :
    a = HeroInfo.iloc[i].str.replace('-','NaN')
    HeroInfo.iloc[i] = a

HeroInfo = HeroInfo.fillna('NaN')
cols = ['Character Score', 'Intelligence', 'Strength', 'Speed','Durability', 'Power', 'Combat']
for c in cols :
    HeroInfo[c] = HeroInfo[c].astype('float')
    HeroInfo[c] = HeroInfo[c].fillna(0)
    
aa = HeroInfo.filter(items=[ 'Character Score', 'Intelligence',
                           'Strength','Speed','Durability','Power','Combat'])
```

# 예제 7-6 추가 메소드 적용하기

누적값 / 누적곱
+ cumsum
+ cumprod

```python
import numpy as np
```

```python
df = pd.DataFrame({'관측값_A':[1,-2,5,3,-1,7],'관측값_B':[3,4,np.NaN,9,2,5]})
df
```

```python
# cumsum() 함수는 원본을 기준으로 각행마다 누적합을 반환한다.
# NaN이 있는 행은 해당 행에서만 누적값을 NaN처리하고 그 이외에는 누적 합을 반환한다.
df.cumsum()
```

```python
# 관측 A 누적값 열 추가 그 데이터는 관측 A 열의 누적합
df['관측값_A_누적값']=df['관측값_A'].cumsum()
df
```

```python
df['관측값_B_누적합']=df['관측값_B'].cumsum()
df
```

```python
# 보기 편하도록 열 재정의
df = df[['관측값_A','관측값_A_누적값','관측값_B','관측값_B_누적합']]
df
```

```python
df = pd.DataFrame({'관측값_A':[1,-2,5,3,-1,7],'관측값_B':[3,4,np.NaN,9,2,5]})
df
```

```python
# insert(n,삽입할 열이름,해당 열 데이터)함수는 n번째 열 인덱스에 열을 삽입한다.
# get_loc(열이름)함수는 열의 수치형 열 인덱스를 반환한다.
# 관측값_A열 다음에 insert하기 위해 1을 더하여 삽입할 열 인덱스를 측정한다.
df.insert(df.columns.get_loc('관측값_A')+1, '관측값_A_누적값',df['관측값_A'].cumsum())
df
```

```python
df.insert(df.columns.get_loc('관측값_B')+1,'관측값_B_누적값',df['관측값_B'].cumsum())
df
```

```python
# 누적곱 : 특정 알고리즘에서 사용한다.
# 누적곱은 원본을 기준으로 누적곱 결과를 반환한다.
# 파이썬에서 오버플로우(숫자 타입 이상의 단위가 발생) 테스트를 하기 위해 다음 예제에서 사용했다
df['관측값_A'].cumprod(), df['관측값_B'].cumprod()
```

```python
df.insert(df.columns.get_loc('관측값_A_누적값')+1,'관측값_A_누적곱',df['관측값_A'].cumprod())
df
```

```python
df.insert(df.columns.get_loc('관측값_B_누적값')+1,'관측값_B_누적곱',df['관측값_B'].cumprod())
df
```

#### 소수 연산 해결법1

```python
import math
```

```python
if math.isclose(0.1+0.2,0.3):
    print("연산 결과 0.3")
else:
    print("0.1 + 0.2는 0.3이 아닙니다!!")
```

#### 소수 연산 해결법2

```python
from decimal import Decimal
Decimal('0.1')+Decimal('0.2')
```

```python
num1 = 0.1
num2 = 0.2
imputation_rate = 10
target = 0.3

num1 *= imputation_rate
num2 *= imputation_rate
target *= imputation_rate

if num1+num2 == target:
    print("연산 일치")
    num1 /=imputation_rate
    num2 /=imputation_rate
    target /=imputation_rate
    print(num1, num2, target)
```
