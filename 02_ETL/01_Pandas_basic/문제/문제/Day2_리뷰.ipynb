# cut 이 very good이고 color가 E이며 가격이 450이상 460인 다이아몬드를 캐럿이 큰 순서대로 나열하시오

```python
diamonds = sns.load_dataset('diamonds')
diamonds[['carat','cut','price']].head(10)
```

```python
diamond2 = diamonds.loc[( (diamonds['color'] == 'E')&(diamonds['cut'] == 'Very Good') & (diamonds['price']<=460) & (diamonds['price']>= 450))]
```

```python
diamond2
```

```python
set_dia = diamond2.set_index('carat')
set_dia.sort_index(inplace=True)
set_dia[::-1]
```
