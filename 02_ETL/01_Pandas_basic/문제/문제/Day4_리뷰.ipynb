# 1. 각 반 별 결석한 학생의 수를 합산하는 열을 추가하고  2. 빠진 학생의 평균을 일주일단위로 보여주는 DataFrame인 week_absent를 만들어라

```python
import pandas as pd
import numpy as np

index=pd.date_range('11/2/2022', periods=14, freq='D')
absent= pd.DataFrame([[1,np.nan],[1,2],[1,1],[np.nan,np.nan],[np.nan,2],[2,1],[3,np.nan],[np.nan,np.nan],[1,2],[2,1],[np.nan,3],[np.nan,2],[2,2],[np.nan,np.nan]]
                  ,index=index,columns=['A반','B반'])
```

```python
absent
```

```python
absent.fillna(0,inplace=True)
```

```python
absent['합산'] = absent.eval('A반 + B반')
```

```python
absent
```

```python
absent_week=absent.resample('W').mean()
```
