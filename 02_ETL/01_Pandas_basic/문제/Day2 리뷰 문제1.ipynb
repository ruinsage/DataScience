```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic
```

### 생존자는 몇 명

```python
(titanic['survived']==1).sum()
```

### 필터링 : 생존한 미성년자 탑승객 정보

```python
under_age = titanic.loc[((titanic['age']<20) & 
                     (titanic['survived']==1)),
                     ['age','class','who','embark_town']]
under_age
```

```python
titanic['survived'].value_counts()
```

```python
survival_rate = titanic['survived'].value_counts(normalize=True) * 100
```

```python
round(survival_rate,1)
```
