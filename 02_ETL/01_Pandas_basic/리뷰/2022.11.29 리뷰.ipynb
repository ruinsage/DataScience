```python
1. 타이타닉호에 탄 클래스별 남여의 평균나이를 구하시오
2. 지역별 남여 생존자 수를 구하시오
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
pdf1 = df.pivot_table(index = 'pclass',    # 행 위치에 들어갈 열
                      columns = 'sex',    # 열 위치에 들어갈 열
                      values = 'age',     # 데이터로 사용할 열
                      aggfunc = 'mean')   # 데이터 집계함수
pdf1
```

```python
pdf2 = df.pivot_table(index='embark_town', 
                      columns='sex',
                      values='survived',
                      aggfunc=['sum','mean'])
pdf2
```

#  문제3

### 이름으로 Inner Join,Left Outer/Right Outer/Outer Join을 하시오

```python
from pandas import Series, DataFrame
```

```python
JDF1 = DataFrame([["홍길동",180,75,1.5],["이순신",160,65,2.0],["이춘복",170,75,0.5]], columns = ["이름","키","몸무게","시력"])
JDF1
```

```python
JDF2 = DataFrame([["홍길동","포워드","잘함","포악함"],["이순신","미드필더","못함","온순함"],["맹꽁이","수비수","잘함","화잘냄"]], columns = ["이름","포지션","실력","성격"])
JDF2
```

```python
pd.merge(left = JDF1 , right = JDF2, how = "inner", on = "이름")
```

```python
pd.merge(left = JDF1 , right = JDF2, how = "left", on = "이름")
```

```python
pd.merge(left = JDF1 , right = JDF2, how = "right", on = "이름")
```

```python
pd.merge(left = JDF1 , right = JDF2, how = "outer", on = "이름")
```
