```python
import pandas as pd
```

### 1. 파일불러오기

```python
book = pd.read_csv('../data/book_data.csv',encoding='cp949')
```

```python
book
```

### 2. 출판사의 카테고리별 판매량을 찾기위해 다른 열들을 빼고 아래와 같이 3개의 열만 남겨두었다

```python
book = book[['book_publisher','book_category','sales_point']]
```

```python
book
```

### 3. sale_point(판매량)을 더해주기 위해서 int type으로 타입변경

```python
book['sales_point'] = book.loc[:,'sales_point'].str.replace(',','').astype('int')
```

```python
book.dtypes
```

### 4. pivot함수를 이용하여 value 값은 'sales_point', index(행)의 값은 카테고리, columns(열)의 값은 출판사로 분리

```python
book = book.pivot_table(values='sales_point', index='book_category', columns='book_publisher', aggfunc='sum')
```

```python
book
```

### 5. NaN 값을 0 으로 처리

```python
book.fillna(0,inplace=True)
```

```python
book
```
