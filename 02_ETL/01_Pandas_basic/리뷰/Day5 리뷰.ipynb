# 문제 1
- taxis DataSet에서 차량별 색깔에 따른 tip에 평균 값을 추출 하시오. ( GroupBy 사용 )

```python
data = {
    "회계연도": ["2017", "2017", "2017", "2018", "2018", "2018", "2019", "2019","2019"],
    "기능": ["일반, 지방 행정", "일반, 지방 행정", "공공질서 및 안전", "공공질서 및 안전", "일반, 지방 행정", "일반, 지방 행정", "공공질서 및 안전", "공공질서 및 안전","공공질서 및 안전"],
    "예산 항목": ['입법 관리', '국정 운영', '재정지원', '정부자원관리', '입법 관리', '국정 운영','재정지원', '정부자원관리','입법 관리'],
    "금액(단위:백만원)": [1135768, 518762, 41003480, 1234123, 6463124, 235511, 2477712, 12677342,326364]
}

df = pd.DataFrame(data,columns=['회계연도','기능','예산 항목','금액(단위:백만원)'])
df
```

# 문제 2
- 상단의 데이터 프레임으로 index = 회계연도, columns = 기능 , values = 금액으로 지정하여 전체 합 피벗 테이블을 작성하고
- 결측값이 발생한다면 0으로 처리 하시오

```python
df1 = df.pivot_table(values='금액(단위:백만원)',columns='기능',index='회계연도',aggfunc='sum')
df1.fillna(0,inplace=True)
df1
```
