```python
import pandas as pd
```

# airport_code 와 country_code를 국가 기준으로 조인한 뒤 정렬한 데이터를 출력하라

```python
airport_code = pd.read_csv('../data/airport_code.csv',encoding='cp949')
```

```python
airport_code
```

```python
country_code = pd.read_csv('../data/country_code.csv', encoding='cp949')
```

```python
country_code
```

```python
country_code = country_code[country_code.columns.difference(['alpha-2'])]
```

```python
code = pd.merge(left = airport_code, right = country_code, how='inner', left_on='국  가', right_on='나라 이름')
```

```python
code
```

```python
code = code[code.columns.difference(['나라 이름'])]
```

```python
code
```

```python
code = code.set_index(['국  가', 'alpha-3', '숫자', '공 항 명'])
```
