# 4. 성별 클래스별 생존자의 합을 구하세요.

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic_sum = titanic.pivot_table(index = 'sex',
                                 columns = 'class',
                                 values = 'survived',
                                 aggfunc = 'sum')
titanic_sum
```

# 7. week 변수의 열인덱스를 행인덱스로 전환하세요.

* 주, 몸무게가 행의 멀티인덱스로 들어가게 하기
