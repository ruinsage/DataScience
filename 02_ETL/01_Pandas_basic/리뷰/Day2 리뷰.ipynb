```python
import seaborn as sns


taxis = sns.load_dataset('taxis')
taxis
```

# 문제1
- taxis DataSet에서 dropoff_zone 열의 값이 Yorkville West 이고 total 값이 20 초과인 택시의 color와 pickup_zone을 추출 하시오

```python
answer = taxis.loc[((taxis['dropoff_zone'] == 'Yorkville West' ) & (taxis['total'] > 20)) , ['color','pickup_zone']]
answer
```

# 문제 2
- taxis DataSet에서 tip이 2.0 이상 이거나 distance가 18.0 초과이고 payment가 cash인 Data의 pickup 시간, dropoff 시간, pickup_zone과 dropoff_zone을 추출 하되
- pickup 시간 별로 정렬하여 추출 하시오

```python
answer2 = taxis.loc[(( (taxis['tip'] >= 2.0 ) | (taxis['distance'] > 18.0) ) & (taxis['payment'] == 'cash')) , ['pickup','dropoff','pickup_zone','dropoff_zone']]
answer2_i = taxis.set_index('pickup')
answer2_i.sort_index(inplace=True)
answer2_i
```
