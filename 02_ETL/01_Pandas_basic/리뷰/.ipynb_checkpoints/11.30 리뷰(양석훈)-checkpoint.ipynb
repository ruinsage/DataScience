```python
import pandas as pd
```

```python
dataframe = pd.DataFrame({'이름' : ['최정','송성문','김성현','김태진','이지영','추신수','푸이그'],
                     '팀명' : ['SSG','키움','SSG','키움','키움','SSG','키움']})
dataframe
```

```python
dataframe2 = pd.DataFrame({'선수명' : ['최정','송성문','김성현','김태진','이지영','추신수','푸이그'],
                            '타율' : [0.476,0.381,0.348,0.333,0.333,0.320,0.261]})
dataframe2
```

```python
record_base = pd.merge(dataframe, dataframe2, left_on = '이름', right_on = '선수명').drop('선수명',axis = 1)
record_base
```

```python
dataframe3 = pd.DataFrame({'이름' : ['최정','송성문','김성현','김태진','이지영','추신수','푸이그'],
                           '안타' : [10,8,8,7,6,8,6],
                           '2루타' : [1,4,1,1,0,1,3]})
dataframe3
```

```python
record_base2 = pd.merge(record_base,dataframe3, on = '이름')
record_base2
```

```python
record_base3 = record_base2.set_index(['팀명','이름'])
record_base3
```

```python
record_base4 = record_base3.unstack(0)
record_base4
```

```python
record_base4.unstack()
```
