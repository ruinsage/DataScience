## 문제2

```python
import pandas as pd
score = pd.read_csv('../data/test_score.csv',index_col='이름', encoding='cp949')
score
```

### 평균점수내기

```python
score['평균'] = score.eval('국어 + 수학 + 사회 + 과학 + 영어')/5
score
```

### 평균 70점 이하 재시험 여부 판단

```python
import numpy as np
mask = score['평균'] < 70
score['재시험여부'] = pd.Categorical(np.where(mask, 'O','X'))
score
```
