```python
base_options = Options()
base_options.add_experimental_option("detach", True)
base_service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=base_service, options=base_options)
# 가상의 크롬브라우저 실행
```

```python
for com_num in range(len(all_rent_page)):
    row_data=[]
    
    url2 = all_rent_page[com_num]
    print(url2)
    driver.get(url2)
    time.sleep(1)
    soup7=BeautifulSoup(driver.page_source,'html.parser')
    time.sleep(1)  
    try:  
        tags6 = soup7.find_all('div', {"class" : "cont_evaluation"})
        time.sleep(1)  
        com_name = soup7.find('h2', {"class" : "tit_location"}).text
        time.sleep(1)  
             
        time.sleep(1)  
        star2 = soup7.find(attrs = {'class' : 'color_b'}).text
        time.sleep(1)  
        print(star2)
        row_data.append(com_name)
        row_data.append(star2)
        print(row_data)
        star_data.append(row_data)

    except:
        tags6 = soup7.find_all('div', {"class" : "cont_evaluation"})
        time.sleep(1)  
        com_name = soup7.find('h2', {"class" : "tit_location"}).text
        time.sleep(1)  
   
        print('별점 없음')
        row_data.append(com_name)
        row_data.append('0점')
        print(row_data)
        star_data.append(row_data)

    
star_data    
```
