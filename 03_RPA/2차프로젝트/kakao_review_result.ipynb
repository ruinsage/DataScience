```python
df = data_frame.iloc[:,:2]
df
```

```python
df['업체명'].unique()
```

```python
df_kakao['RENTAL_CAR_ID'].unique()
```

```python
df2 = pd.merge(df, df_kakao, how='inner', left_on = '업체명', right_on = 'RENTAL_CAR_ID')
df2
```

```python
df3 = df2.drop(['업체명','RENTAL_CAR_ID'],axis=1)
df3
```

```python
df4 = df3.rename(columns={'id':'RENTAL_CAR_ID'})
df4
```

```python
df5 = df4[['WRITER','CONTENT','TYPE','SCORE','CREATED_AT','RENTAL_CAR_ID']]
df5
```

```python
df6 = df5.copy()
```

```python
df6.reset_index(inplace=True)
```
