```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('2022_렌터카불편사항.csv', encoding='utf-8')
df.rename(columns={'Unnamed: 0':'구분'}, inplace=True)
df.rename(columns={'Unnamed: 1':'상세'}, inplace=True)
```

```python
df2 = df[df['구분'] == '연령']
df99 = df2[['상세','비싼 가격', '예약 불편/어려움', '원하는 차량 없음', '차량(점검) 부실',
       '렌터카 호객행위', '렌터카 회사의 불법주차(공터주차)', '렌터카 회사의 불친절', '취소/환불 불편/어려움', '기타',
       '불편/불만 없음']]
df99.index = df99['상세']
df99 = df99.sort_index(ascending=False)
df99.plot(kind='barh', figsize=(10,6), stacked=True, cmap='coolwarm')
title_font = {
    'fontsize': 18,
    #'fontweight': 'bold'
}
plt.title('22년도 연령별 렌터카 불만사항 비율', fontdict=title_font, loc='left', pad=30)
plt.legend(loc=(1.01,0))
plt.ylabel('연령별')
x = np.arange(0,110,10)
x
plt.xticks(x)
plt.show()
```
