```python
import folium
import pandas
import requests
import json
```

```python
url = 'http://api.jejuits.go.kr/api/infoRoadEventList'
key = 860553
```

```python
def get_request_url(url):
    params = {'code': key}
    response = requests.get(url, params=params)
    return response.text
```

```python
def json_to_df_wave(raw_json):
    all_data = []
    column_list = ['ocrr_id', 'latitude', 'longitude', 'heading', 'link_id', 'code']

    for record in raw_json['info']:
        row_data = []
        for column_data in column_list:
            row_data.append(record.get(column_data))
        all_data.append(row_data)

    column_list = ['ID', '위도', '경도', '방향', '노드', '코드']
    return column_list, all_data
```

```python
def info_collector():
    raw_str_json = get_request_url(url)

    if raw_str_json:
        raw_json = json.loads(raw_str_json)

    column_list, all_data = json_to_df_wave(raw_json)

    df = pandas.DataFrame(all_data, columns=column_list)
    return df
```

```python
data = info_collector()
data
```

```python
map = folium.Map(location=[33.21, 126.32],zoom_start=10)
coords = data[['위도','경도','코드']]
for lat, long, code in zip(coords['위도'], coords['경도'], coords['코드']):
    folium.Marker([lat, long], tooltip=code, icon = folium.Icon(color="black", icon="star")).add_to(map)
map
```
