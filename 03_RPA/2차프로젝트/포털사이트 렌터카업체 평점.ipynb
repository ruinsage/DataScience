```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
input_file = '제주도_렌터카업체위치 (1) (2).csv'
output_file = 'kakao_review_result.csv'
df_com = pd.read_csv(input_file, encoding = 'cp949')
df_com
```

```python
area = df_com['자동차총보유대수']/10
```

```python
plt.rcParams['figure.figsize'] = (10, 10)
#plt.rcParams['font.size'] = 12

title_font = {
    'fontsize': 18
}

x = df_com['id']
y1 = df_com['네이버 평점']
y2 = df_com['구글 평점']
y3 = df_com['카카오 평점']

fig, ax1 = plt.subplots()

ax1.scatter(x, y1, color='#3dd0d3', label = '네이버', marker= 'd')#, s=area)
ax1.set_ylim(0.8, 5.2)
ax1.set_xlabel('업체 id')
ax1.set_ylabel('평점')
ax3.set_yticks([])

ax2 = ax1.twinx()
ax2.scatter(x, y2, color='#dc9173', label = '구글', marker= '*')# , s=area)
ax2.set_ylim(0.8, 5.2)
ax2.set_yticks([])

ax3 = ax1.twinx()
ax3.scatter(x, y3, color='#ebc06e', label = '카카오', marker= 'v')#, s=area)
ax3.set_ylim(0.8, 5.2)
ax3.set_yticks([])

ax1.legend(loc=(1.01,0.0801))
ax2.legend(loc=(1.01,0.0401))
ax3.legend(loc=(1.01,0.001))

plt.title('포털 사이트 렌터카업체 평점',fontdict=title_font, loc='left', pad=30)

plt.show()
```
