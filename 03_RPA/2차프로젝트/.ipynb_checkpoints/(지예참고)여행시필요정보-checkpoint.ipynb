```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('2018_2022_여행필요정보.csv', encoding='ms949')
df.rename(columns={'Unnamed: 0':'구분'}, inplace=True)
df.rename(columns={'Unnamed: 1':'상세'}, inplace=True)
```

```python
df[df['구분'] == '연령'].columns
```

```python
sel_year = 2022
df18 = df[df['년도'] == sel_year]
df18 = df18[df18['구분'] == '연령']
df99 = df18[['상세','방문지 정보(개장시간, 입장료, 여행코스 등)','이동 거리 및 교통편','음식 및 맛집 정보',
             '쇼핑 장소','숙박시설','지역 축제 및 행사정보', '역사/문화유적/전통문화 체험 관련 정보',
             '금융 정보(ATM, 환전소 등)','치안, 응급상황(병원, 경찰 등)','기타','코로나 확진자 수/동선',
            '병원/코로나 진료소', '코로나 방역']]
df99.index = df99['상세']
df99 = df99.sort_index(ascending=False)
df99.plot(kind='barh', figsize=(10,6), stacked=True, cmap='BuPu')
title_font = {
    'fontsize': 18,
    #'fontweight': 'bold'
}
plt.title(f'{sel_year}년도 연령별 필요정보 비율', fontdict=title_font, loc='left', pad=30)
plt.legend(loc=(1.01,0))
plt.ylabel('연령별')
x = np.arange(0,110,10)
x
plt.xticks(x)
plt.show()
```
