```python
import urllib.request
from bs4 import BeautifulSoup
import re
import urllib.request as ur
import pandas as pd
```

```python
html = urllib.request.urlopen('http://www.jejutori.com/CS/TourInfo/detail.aspx?cid=4808')
soup=BeautifulSoup(html, 'html.parser')
soup
```

```python
tags = soup.find_all('td', {"class" : "NHN_Layout_Preset"})
tags
```

```python
corse=re.compile('[<]p\s+align=["]justify["][>][<]?b?r?[/]?[>]?[\r]?\s*\d*[.]\s(.*)[<][/]p[>]')
c=corse.findall(str(tags))
print(c)
```
